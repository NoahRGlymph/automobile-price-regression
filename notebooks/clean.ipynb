Data Cleaning:

In this notebook I prepare the car dataset for cleaning.

The goal is:

-handle missing values

-convert variables to the proper data types

-create a clean dataset that can be used for modeling later on


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/imports-85.data", header=None, na_values="?")
```

Missing Column Names:

There were missing variables and the original data set didn't even have column names. I labeled the columns so I could figure out which pieces of data corresponded with which variables.

```python
column_names = ["symboling","normalized_losses","make","fuel_type","aspiration","num_doors","body_style","drive_wheels","engine_location","wheel_base","length","width","height","curb_weight","engine_type","num_cylinders","engine_size","fuel_system","bore", "stroke","compression ratio", "horsepower","peak_rpm","city_mpg","highway_mpg","price"]
df.columns = column_names
```

Missing Variables:

Normalized losses and number of doors had some missing values. Numerical values were replaced using the median of the column, categorical values using mode.

```python
df["normalized_losses"].fillna(df["normalized_losses"].median(), inplace=True)
df["num_doors"].fillna(df["num_doors"].mode()[0], inplace=True)
df["price"].fillna(df["price"].mode()[0], inplace=True)
df["bore"].fillna(df["bore"].mode()[0], inplace=True)
df["stroke"].fillna(df["stroke"].mode()[0], inplace=True)
df["horsepower"].fillna(df["horsepower"].mode()[0], inplace=True)
df["peak_rpm"].fillna(df["peak_rpm"].mode()[0], inplace=True)
```

Verifying Missing Values:

Missing values have been removed, lets check thats true.

```python
df.isna().sum()
```

Export Clean Dataset:

After cleaning the dataset, it is exported as a new .csv file.

```python
df.to_csv("../data/clean/automobile_clean.csv", index=False)
```
